# History and Time Travel
1. Query Delta Lake table history.
2. Query previous versions of the data.
3. Query data from a specific time.
4. Restore data to a specific version.

```python
%sql
DESC HISTORY demo.delta_lake.companies;
```

```python
%sql
SELECT * FROM demo.delta_lake.companies;
```

```python
%sql
SELECT * FROM demo.delta_lake.companies
VERSION AS OF 1;
```

```python
%sql
SELECT * FROM demo.delta_lake.companies
TIMESTAMP AS OF '2025-10-17T06:51:45.000+00:00';
```

```python
%sql
RESTORE TABLE demo.delta_lake.companies
VERSION AS OF 1;
```

```python
%sql
SELECT * FROM demo.delta_lake.companies;
```

```python
%sql
DESC HISTORY demo.delta_lake.companies;
```
