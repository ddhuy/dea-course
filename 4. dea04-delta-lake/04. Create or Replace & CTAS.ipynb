## 1. Difference between Create or Replace and Drop and Create Table statements

```python
%sql
DROP TABLE IF EXISTS demo.delta_lake.companies;


CREATE TABLE demo.delta_lake.companies (
  company_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  company_name STRING,
  founded_date DATE,
  country STRING
);


INSERT INTO demo.delta_lake.companies (company_name, founded_date, country)
VALUES
  ('Acme Inc', '2010-01-01', 'USA'),
  ('Beta Corp', '2015-05-01', 'Canada'),
  ('Gamma Inc', '2018-08-01', 'USA'),
  ('Delta Corp', '2020-03-01', 'USA'),
  ('Apple', '1976-04-01', 'USA'),
  ('Microsoft', '1975-04-01', 'USA'),
  ('Google', '1998-09-01', 'USA'),
  ('Amazon', '1994-05-01', 'USA'),
  ('Facebook', '2004-02-01', 'USA'),
  ('Tesla', '2003-07-01', 'USA');
```

```python
%sql
DROP TABLE IF EXISTS demo.delta_lake.companies;
```

```python
%sql
CREATE OR REPLACE TABLE demo.delta_lake.companies (
  company_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  company_name STRING,
  founded_date DATE,
  country STRING
);


INSERT INTO demo.delta_lake.companies (company_name, founded_date, country)
VALUES
  ('Acme Inc', '2010-01-01', 'USA'),
  ('Beta Corp', '2015-05-01', 'Canada'),
  ('Gamma Inc', '2018-08-01', 'USA'),
  ('Delta Corp', '2020-03-01', 'USA'),
  ('Apple', '1976-04-01', 'USA'),
  ('Microsoft', '1975-04-01', 'USA'),
  ('Google', '1998-09-01', 'USA'),
  ('Amazon', '1994-05-01', 'USA'),
  ('Facebook', '2004-02-01', 'USA'),
  ('Tesla', '2003-07-01', 'USA');
```

## 2. CTAS statement

```python
%sql
CREATE OR REPLACE TABLE demo.delta_lake.companies_canada
AS
  SELECT
    CAST(company_id AS INT) AS company_id,
    company_name,
    founded_date,
    country
  FROM demo.delta_lake.companies
  WHERE country = 'Canada';

SELECT * FROM demo.delta_lake.companies_canada;
```

```python
%sql
DESC demo.delta_lake.companies_canada;
```

```python
%sql
ALTER TABLE demo.delta_lake.companies_canada
  ALTER COLUMN founded_date COMMENT 'Date the company was founded';
```

```python
%sql
ALTER TABLE demo.delta_lake.companies_canada
  ALTER COLUMN company_id SET NOT NULL;
```

```python
%sql
DESC EXTENDED demo.delta_lake.companies_canada;
```

```python
%sql
DESC HISTORY demo.delta_lake.companies_canada;
```
